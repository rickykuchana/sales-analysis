```python
import pandas as pd

df = pd.read_csv("sales.csv")
df
```

```python
df["revenue"] = df["quantity"] * df["price"]
df
```

```python
region_revenue = df.groupby("region")["revenue"].sum()
print("Revenue by region:")
print(region_revenue)
```

```python
category_revenue = df.groupby("category")["revenue"].sum()
print("Revenue by category:")
print(category_revenue)
```

```python
product_revenue = df.groupby("product")["revenue"].sum().sort_values(ascending=False)
print("Top products by revenue:")
print(product_revenue)
```

```python
average_order = df["revenue"].mean()
print("Average order value:")
print(average_order)
```

```python
print("Biggest order:")
print(df["revenue"].max())
print("Smallest order:")
print(df["revenue"].min())
```

```python
orders_per_region = df.groupby("region")["order_id"].count()
print("Orders per region:")
print(orders_per_region)
```

```python
units_per_product = df.groupby("product")["quantity"].sum().sort_values(ascending=False)
print("Units sold per product:")
print(units_per_product)
```

```python
big_orders = df[df["revenue"] > 1000]
print("High-value orders over $1000:")
print(big_orders)
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["month"] = df["order_date"].dt.month
monthly_revenue = df.groupby("month")["revenue"].sum()
print("Revenue by month:")
print(monthly_revenue)
```

```python
print("=== SALES SUMMARY ===")
print("Total revenue:", df["revenue"].sum())
print("Total orders:", len(df))
print("Average order value:", df["revenue"].mean())
print("Number of regions:", df["region"].nunique())
print("Number of products:", df["product"].nunique())
```
